## 2.6.1 Pass by reference (vs pass by value)

Practice the next program, and check out the results.

````python
listA=["orange","apple","banana"]
listB=listA
print(listB)
listA[0]="grape"
print(listB)
listB[1]="tomato"
print(listA)
````

A list `listA` of 3 items is defined at Line 1 and then copied to `listB` at Line 2. `listB` is printed at Line 3, which shows of course  the same content as `listA`. So far it is natural, isn’t it?

The problem is from  line 4 and after.
At Line 4, the first items of `listA` is replaced with “`grape`”. Looking at the print of line 5, we observe that `listB[0]`, the first items of `listB`, is also replaced with “`grape`”.

Whlile only `listA[0]` was changed,  `listB[0]` is also changed synchronously. Why?
It is because there is a secret when  `listA` is copied to `listB` in line 2. Actually, the copy of lists is actually a “pass by reference” (the reference to `listA` is passed to `listB`). Plainly said:


> `listB` does not only take over the content of `listA`, but also its destinity is linked to  `listA`.

`listB` is not only another list which have the same content as `listA`, but now any action on the items of either lists will similarly affects the other list.

On this way, listA and listB are "in the same boat", they built up a relationship that implies that if  we operate on `listB`, `listA` is changing in accordance. In line 6, `listB[1]` is revised to “`tomato`” and  we can observe that `listA[1]` has changed to “`tomato`” too.


The mechanism inside the computer that explains this behavior is simple: `listA` and `listB` are pointing to the same position of the memory. This memory space is storing the content (the items) of `listA`. When we copied `listA` to `listB`, this was not that memory (the content of `listA`) that was copied, but the reference to it, whence the name "pass by reference".



## 2.6.2 Pass by value (vs pass by reference)

To copy a list without being linked by fate to the original list, it suffices to copy the *content* of the list (and not its reference only). This kind of copy is called “**pass by value**”, as opposed to "pass by reference".



````python
listA=["orange","apple","banana"]
listB=listA[:]
print(listB)
listA[0]="grape"
print(listB)
````

We revised a little the previous program. We only changed the assignment of line 2:

```
listB=listA[:]
```

The right side uses slice notation, but the beginning and the end of the slice are both omitted. This notation means the "whole of list" (without delimitation).  Therefore, the meaning of this assignment is

> the whole of `listA` is assigned to `listB `.

This copy is different from the previous one which was passed by reference. Only the content is copied, that is "values are passed" from one list to another. Notice the result displayed at Line 5: `listB[0]` is still "orange", it has not become "grape".

The following program also copy a list as pass by values:




````python
listA=["orange","apple","banana"]
listB=[w for w in listA]
print(listB)
listA[0]="grape"
print(listB)
````

Description of Line 2:

```
listB=[w for w in listA]
```
This method called "**list comprehensions**" creates a list. The meaning of the right side is

* `w for w in list A` means "each item w which belongs to `listA` ".

* By enclosing the above with square brackets [], it becomes the list made of all such `w` (not more a list comprehension).
The resulting list is assigned to `listB`, as a result it  realizing a copy pass by value of `listA`.



````python
list_comp=[x**2 for x in [1,2,3,4]]
print(list_comp)

another=["a"+word for word in ["b","c","d"]]
print(another)
````

## 2.6.3 Copy of a dictionary

Notice that the copy of standard variables is pass by values, while copy of lists is pass by reference.

What about the copy of dictionaries? The conclusion is that it is pass by reference, of course. In the next example, we can see that the copy of dictionary is  pass by reference. (Note the result displayed in line 5)


````python
dictA={"name":"Koyuza", "age":73, "home":"Yamanashi"}
dictB=dictA
print(dictB)
dictA["age"]=74
print(dictB)
````

To perform a pass by value copy of a dictionary, we do not make a simple assignment, but call the `copy` instruction as in the following example. (Note the line 2)

````python
dictA={"name":"Koyuza", "age":73, "home":"Yamanashi"}
dictB=dictA.copy()
print(dictB)
dictA["age"]=74
print(dictB)
````

In the output of line 5, "age" is still 73, so we can observe that the copy is pass by value.

## 2.6.4 Tuple Type

By using a list, we are able to handle a variable that has multiple “items”. There is a data type very close to the list type called “Tuple type”.
Both *list* and *tuple* have multiple items, *tuple* differs from *list* by:

* Instead of using square brackets [ ],  round brackets ( ) are used.
* Items cannot be changed: no deletion, no insertion, no changes.

Especially because of the second restriction, we might feel that it is more inconvenient to use *tuple* than *list*. However, tuple is lighter than *list* (less burden inside the computer processing). There are advantages that it can prevent unexpected mistakes related to updating and increasing the items.

When we want to define a data structure that can contain several items, but don't need to operate on them, then using *tuple* is a more suitable choice than *list*.


````python
a=("orange","banana","apple")
print("I like",a[0],"but not",a[1])
a[2]="tomato"
````

````python
test=(1,"word",1+1j)
print(test)
````

This is an example of using tuple. Tuple `a` which has 3 items is defining in line 1. It’s the same definition method as list that others use round bracket ( ) to replace square bracket [ ]. 

In line 2, we are referencing items of `a`, and the way it is accomplished is the same as for list. `a[0], a[1], a[2]` are included in round bracket.

In line 3, there is an error in this program. The content of `a[2]` is revised according to the assignment  in line 3. But tuple cannot do content change of items after something once been defined. In this way, items which were generated by tuple are like "specific variable for reading". This is the main difference from list.

##2.7.3 (Question 3)
Pass-by-value, pass-by-reference is quite tricky.
What is the difference betweem the two below?

````python
# Code 1
L=[[1, 2], [4, 5]]
R=L[:]
R[1][0]=0
print(L[1][0])
````

````python
# Code 2
L=[[1, 2], [4, 5]]
R=L[:]
R[1]=[0,0]
print(L[1][0])
````

Looking at line 3 of both snippets, it seems like they are all using pass-by-value (further denoted as PBV heretofore) to assign the values of L to R. We would then instinctively think that we can change the values of R however we want and the list L would never change. This isn't true, because as we can see in line 4, there's a difference in how the value of R[1][0] is changed. In snippet 1, we are accessing R[1][0] directly, which is NOT a list. This means we are changing what the value itself points to, and as the values in the inner lists of R and L are still pointing to the same memory space, the element with the same position in L also changes.

Contrast this to the second snippet, where we only point to R[1] and changing the inner list R[1] itself. This means we are not changing what R[1][0] points to, we are changing its value by essentially disassociating it with R[1], and then assigning the new value [0.0] to R[1]. This is PBV, which will not change what L[1] or L[1][0] is.
