# 4. Iterations

When designing algorithms, it is common to include operations that repeat the same process as long as a certain condition is satisfied. We call this repetition of process an "**iterations (loops)**".This is one of the basic syntax, along with conditional branching that we learned in the previous section, and is used not only in Python but other programming languages as well.

By combining pieces of programs that use iterations and conditional branching, we can write more complex programs.

# 4.1 Iterations using the `for` statement

In Python, the syntax used to describe iterations is the `for` statement. Other programming languages, such as C and Java, also use the `for` statement, but the syntaxes differ slightly in each language.



## 4.1.1 Basics of `for` statements

Let us look at an example that uses the `for` statement.

````python
print("input a number: ")
n=int(input())
sum=0

for x in range(n+1):
  sum=sum+x 
  print("sum =", sum)
````

This program calculates the sum of all natural numbers from 1 to n for a given natural number n, and prints out the result. If we use the summation formula of an arithmetic sequence, we can simply calculate $n(n+1)/2$ and print out the result. However in this program, we calculate `sum` by adding 1, 2, 3, 4, ..., n one by one, and print out the result of `sum` after each addition. (The final result is the sum of the numbers 1 to n.)

The program does the following steps.

* First, prepare a variable `sum` and assign 0 to it (3rd line).
* In the order x=1,2,3,... and so on, until x=n, x is added to `sum` (and the result is printed out each time).

This iteration is executed by the `for` statement in the 5th line to the  7th line.




Before explaining `for` statemtents (or `for` loops), let us discuss about `range`.
(More details about `range` will be discussed later.)
The basic use of `range` is as follows.

```
range(n)
```

A single integer value `n` is specified in the parentheses.
This means that the `range` is "integers from 0 to the number before n", i.e. "integers from 0 to n-1".
Note that `n` is not included in the `range`.


Now, let us move on to the explanation of the `for` loops.
Lines 5 to 7 in the example above are written in the following form.

```
for x in range(n+1):
  [Loop body A]
```

The statement following `for`, which is

```
x in range(n+1)
```

means "each integer `x` in the range specified by `range(n+1)`".
As mentioned above, `range(n+1)` is the range from 0 to n, and n+1 is not in the `range`.

For each integer x in this range, `Loop body A` is repeated in the order of x=0,1,2,... up until n.
(**Similar to the  `if` statement, `Loop body A` should be indented. Inside Colab, this is done automatically after you press Enter, otherwise press the [Tab] key**)
In other words, the statement in the 5th line means

> Execute the iteration of `Loop body A` in the order of x=0,1,2,...,n.

In this example, lines 6 and 7 correspond to `Loop body A`. The assignment statement in line 6 means "add `x` to `sum`", and line 7 shows the result of `sum` after adding `x`.

By doing this, starting from `sum=0`, `x` is changed in the order of x=0,1,2,3,. . n, and added repeatedly to `sum`

## 4.1.2. Specifying a loop variable using `range()`

As mentioned in the example above, the `for` statement combined with `range()` is the most basic statement for iterations, and can be used to increase the value of x from 0 (or 1) one by one, while repeating certain operations until x reaches a certain value.
Here, we will call the variable x a "**loop variable**".

````python
print("input five numbers: ")
sum=0

for x in range(5):
  sum=sum+int(input())

print("Sum =",sum)
````

In the example above, five integers are input from the keyboard and the sum of the numbers is calculated.
The `for` statement in the 4th line is followed by `range(1,6)`, and the loop body, which is the 5th line, is executed sequentially for x=1,2,3,4,5.


````python
print("input a number n: ")
n=int(input())
sum=0

for x in range(0,n+1,2):
  sum=sum+x

print("Sum of even numbers up to",n,"is",sum)
````

The example above is a program that calculates the sum of all even numbers from 1 to n (the integer n is entered in the 2nd line) and prints out the result.
In the 5th line, the loop variable `x` is specified by `range(0,n+1,2)`, which is an integer value from 0 to n with increments of 2. As a result, `x` will only use even numbers from 0 and n.

**Supplement.** Look at the following example.

````python
for x in range(5):
  print("x =", x)
  x=0

print("x =", x, "after for loop")
````

In this example, the loop variable `x` is reset to 0 at the end of the loop body (the 3rd line).
In this case, it seems that the loop will never end, because the value of the loop variable `x` will remain as 0 forever.

However, if you run the program, you will see that the program works fine and `x` will be printed out as 0, 1, 2, 3, and 4.
Even if you manipulate the loop variable `x` in the loop body, the number of loops itself does not change.

So, the process "If a certain condition is met while executing a loop, reset `x` to 0 and start over" is not possible with a `for` loop.
In such a case, you should use another iteration syntax.

Looking at the results after the `for` loop is completed (which is the 5th line), the effect of the 3rd line remains, making `x=0`.

## 4.1.5 Stopping loops by using `break`

While executing a `for` loop, we might want to stop the loop when certain conditions are met. In such a case, we can use the `break` statement to forcibly break out of the loop.

````python
print("input a number n: ")
n=int(input())
sum=0

for x in range(1,n+1):
  sum = sum + x
  if sum>=1000:
    break

print("Sum =",sum, " after ", x, " iterations" )
````

In this example, for the natural number n that is input, the sum of the natural numbers from 1 to n is calculated. If the sum reaches 1000 or more during the calculation, the process is terminated and the value of the sum at the time of termination is displayed.

The `for` loop contains an `if` statement (line 7), which determines whether the current `sum` is greater than or equal to 1000. If `sum` is greater than 1000, the `break` statement in line 8 forces the `for` loop to stop. If `sum` has not reached 1000, the `for` loop is continued.

**Supplement: The value of the loop variable after the `for` loop ends**

As you can see in the result of the 4th line in the following example, after the `for` loop finishes, the loop variable `x` has the value 4, which is the value of the end of `range(5)`.

````python
for x in range(5):
  print("x =",x)

print("x =",x,"after for loop")
````

Let us look at the following example, where `break` occurs in the middle of a `for` loop.
In this example, `x` is supposed to go from 0 to 9, but the 2nd and 3rd lines force the loop to stop with a `break` statement when `x` is greater than or equal to 5.
Thus, the value of x is printed as 5 in the 6th line.

````python
for x in range(10):
  print(x)
  if x>=5:
    break
  x=7

print("x =",x)
````

Note that the assignment statement `x=7` is found in the `for` loop in the 4th line, but it has no effect.

# 4.2 Iterations using `while` statements

Besides the `for` statement, there is another syntax for describing iterations, which is the `while` statement.

## 4.2.1 Basics of `while` statements

The program below asks for an interger $n\ge 1$ and doubles it iteratively  $2n,\ 4n,\ 8n,\ldots$ while the largest double is $< 1000$. The first double Line 4 is outside the `while` loop, since otherwise the first print Line 6 will display $n$ instead of its double.

````python
print("input a number greather than 1 ")
n=int(input())

n=2*n
while n<1000:
  print("double is ", n)
  n=2*n
````


`while` statements are written in the following form.

```
while [Condition A]:
  [Loop body B]
```

Similar to `for` statements, indent the part of `Loop body B` (done automatically in Colab, otherwise press the [Tab] key).
In this syntax, the process written in `Loop body B` is repeated as long as `Condition A` is satisfied. The process works as follows.

* First, decide whether `Condition A` is True or False. If `Condition A` is False, the iteration is stopped.
* If `Condition A` is True, `Loop body B` is executed in one step. Then, return back to decide again whether `Condition A` is True or False.

##4.2.2 `for` loops and `while` loops

Let us look at an example where a `for` loop and a `while` loop do the same operation.
The two programs both compute the sum of 1 to n for the input n and print out the result.

````python
print("input a number n: ")
n=int(input())
sum=0

for x in range(1,n+1):
  sum=sum+x

print("Sum =",sum)
````

````python
print("input a number n: ")
n=int(input())
sum=0

x=1
while x<=n:
  sum=sum+x
  x=x+1

print("Sum =",sum)
````

The first program uses `range()` in the `for` loop, the loop variable `x` runs through 1 to n.

On the other hand, in the second program, the loop variable `x` is initialized to 1 in the 5th line before entering the `while` loop in the 6th line, and `x` is increased by 1 in the 8th line during the loop. Since the condition to continue the iteration is "`x` is less than or equal to `n`", this will result in x going from 1 to n.

The difference between `for` and `while` statements is as follows.

* The `for` loop is suitable for iterations where the value of the loop variable is predetermined.
* The `while` loop is suitable where the number of iterations is indefinite and the value of the loop variable may be manipulated during the process depending on the situation.

In the following example, a `while` loop is used to terminate the process when the input value from the keyboard is a negative number.

````python
print("positive numbers, please: ")
x=0

while x>=0:
  x=int(input())

print("Oops! You give a negative number",x)
````

##4.2.3 Infinite loop

Depending on `Condition A`, the `while` loop may run forever. For example, let us look at the following program.


```
x=0

while x<=10:
  x=x-1
```

In this example, we start with `x=0` and repeat the loop as long as `x` is less than or equal to 10.
However, each time the loop is executed, the value of `x` decreases by 1, so the change in the value of x is x=0, -1, -2, -3, ... and the condition `x<=10` will always be True. As a result, the `while` loop becomes an *infinite loop* that will never stop.


##4.2.4 Loops that are never executed.

Sometimes, the loop body of a `while` loop ends up never being executed. For example, let us look at the following example.





````python
x=11

while x<=10:
  x=x+1

print(x)
````

In this example, the condition for executing the `while` loop in lines 3 to 4 is `x<=10`, but the value of `x` is 11 just before entering the `while` loop.
Therefore, when `Condition A` (the 3rd line) is first decided, it is False, and as a result, `Loop body B` (the 4th line) is not executed at all, and the `while` loop ends.

As a result, Line 6 will display the initial value of `x`, which is 11.

##4.2.5 `do~while` loops

If you have experience in other programming languages such as C or Java, you may know that there is a syntax for iterations similar to the `while` loop, called the `do~while` loop.
Its structure is shown in the following figure.

<img src="http://drive.google.com/uc?export=view&id=1TkdCHBf9Ppi5g7zqCLcalRbgZ6fNk5ph" height="200">

* First, `Loop body B` is executed once.
* Next, determine whether `Condition A` is True or False. If `Condition A` is True, it executes `Loop body B` once, and then returns to the decision making of `Condition A`.
* If `Condition A` is False, the loop is terminated.

It is similar to the `while` loop, but the difference lies in the order of execution of the loop body and the decision making.

* In the `while` loop, determine the condition → execute the loop
* In the `do~while` loop, execute the loop → determine the condition.

For this reason, unlike the `while loop`, in the `do~while` loop, `Loop body B` is executed at least once.

However, Python does not have a syntax for `do~while` loops.
This is because the `do~while` loop can be realized by using `while` and `break` statements.

````python
print("positive numbers, please: ")

while True:
  x=int(input())
  if x<0:
    break

print("Oops! You give a negative number",x)
````

In this example, the program will stop when the value entered from the keyboard is a negative number.
The 3rd line is

```
while True:
```

This statement assigns the value `True` to `Condition A`.
This means that we have specified that `Condition A` is always True, so the `while` loop will never end.

However due to the `if` statement in lines 5 to 6, when the value `x` entered in the 4th line is negative, the `break` statement in the 6th line forces the `while` loop to stop.

The 4th line is always executed at least once, and then the `if` statement in lines 5 and 6 checks the condition "is `x` greater than or equal to 0?"
With this, the `do~while` loop is realized.


# 4.3 Nesting iterations

It is also possible to construct a loop inside another loop.
Let us see an example of this in the following program. 

````python
print("input a number n: ")
n=int(input())
count=0

for x in range(2,n+1): # outer loop
  d=2
  
  while d*d<=x: # inner loop
    if x%d==0: # x%d means "reminder of the division of x by d"
      break
    else:
      d=d+1
# end of the inner loop
  if d*d>x:
    count=count+1  # if we reach this line, x is a prime number
# end of the outer loop

print("There are",count,"primes up to",n)
````

For the natural number `n` entered in the 2nd line, this program counts the number of primes from 2 to `n`, and outputs the result. The variable `count` records the number of prime numbers found so far, and its initial value is of course 0 (line 3).

Lines 5 to 15 are part of a `for` loop (outer loop), and a `while` loop (inner loop) consisting of lines 8 to 12 is embedded in it.

<img src="http://drive.google.com/uc?export=view&id=1eRu9fZ_hDE7F5bYIvS3sNYqnSsaS7WXo">

The `for` outer loop determines for each `x` from 2 to n, whether `x` is prime or not, and if `x` is prime, it increases the value of `count` by one. The `while` loop inside is responsible for deciding whether each `x` is a prime number or not.

The `while` inner loop, starting from `d=2` and increasing `d` by 1, decides whether `x` is a divisor of `d`. The `if` statement (line 9) in the `while` loop decides if `x%d==0` holds, which means "`x` is divisible by `d`" or "`d` is a divisor of `x`".
The candidate for the divisor `d` can range from 2 to the square root of `x`, $\sqrt{x}$. (Thus, the condition of the while loop is `d*d<=x`.)
This is based on the fact that

* If `x` is a composite number (not a prime number), then there must be a divisor of `x` in the range from 2 to $\sqrt{x}$.
* If x is prime, then there is no divisor of `x` in that range

When `d` moves from 2 to $\sqrt{x}$ in sequential order,

* If `x%d==0` is True, then `x` is confirmed to be a composite number with divisor `d`, and the `break` statement in the 10th line terminates the `while` loop at this point. (**The `break` statement terminates the `while` inner loop, but not the `for` outer loop**.)
* If `x%d==0` is False, then `d` is not a divisor of `x`, so we increase `d` by one and return to the beginning of the `while` loop, as shown in lines 11 to 12.

The `while` loop is exited in one of the following two cases.

1. When a divisor `d` of `x` is found during the loop and the `break` statement in line 10 forces the loop to terminate. This happens when `x` is a composite number, in that case we leave the `while` loop even though `d*d<=x` is True.
2. When `d` exceeds $\sqrt{x}$ and `d*d>x`. This happens when `x` is a prime number.

Therefore, we can tell whether the `while` loop ended in 1. or 2. from the relationship between `d*d` and `x` at the end of the `while` loop.

We use the `if` statement in line 14 to determine whether `d*d>x` is True or False, and if it is True, we increase `count` by 1 since x is prime (line 15).
This is the content of the `for` loop.

After exiting the `for` loop, the final `print()` function (line 17) prints out the value of `count`, that is, the number of prime numbers found.

## 4.4.1 List comprehension

We can use a method called **list comprehension** (already touched in Section 2.6.2) to define a list. This is simply a way to define a list in the form of "a list consisting of △△ that satisfies conditions 〇〇". Let us look at a simple example.

````python
list=[x for x in range(10)]
print(list)
````

The 1st line defines `list` using list comprehension method. In `[ ]`, the latter part 

```
for x in range(10)
```

means "each `x` specified in `range(10)`". That is, `x` is an integer between 0 and 9.

And by putting `x` in front of `for`, the 1st line means "to create a list consisting of all such x".

Next, let us slightly rewrite the 1st line as follows.

````python
list=[x+2 for x in range(10)]
print(list)
````

The `x` in front of `for` has been replaced by `x+2`. As you can see from the result of the 2nd line, we are now told to make a list of `x+2` for each `x` specified in `range(10)`.

In other words, the part after `for` specifies the range of `x`, and the part before `for` is an expression related to `x`, and a list is made using the values calculated by the expression.

Also, if you put an expression that is irrelevant to `x` before `for`, as in the following example, a list is created with that value as its items.

````python
list=["a" for x in range(10)]
print(list)
````

## 4.4.2 Extracting indices in lists

Let us continue to talk about lists.
If we want to know the index of an item in a list, we can use the `index()` function as shown in the following example.

````python
a=["orange","apple","banana","grape","apple","tomato"]
print(a.index("apple"))
````

In the 2nd line,

```
a.index("apple")
```

refers to the index of the item with the value `apple` in list `a`.

If you look closely at the 1st line, you will see that `a` has two indices that has the value `apple` (`a[1]` and `a[4]`).
The `index()` function only returns the index of the first `apple` found, i.e., 1.

In the next example, we want to get all the index numbers of the item `apple`.

````python
a=["orange","apple","banana","grape","apple","tomato"]
b=[]

for x in range(len(a)):
  if a[x]=="apple":
    b=b+[x]

print(b)
````

In this example, the list `b` will contain all indices k such that `a[k]="apple"`. The initial value of b is an empty list as shown in the 2nd line.

The for loop in line 4 runs through all indices of list `a` in the order of 0,1,2,... 
(`len(a)` is the length of list `a`).
For each x, line 5 determines whether `a[x]` is `apple` or not, and if it is, line 6 appends `x` to list `b` as an item.

Another way is to use the method of list comprehension.
To do so, we use the **`enumerate()` function**.
`enumerate()` is used in the following form.

```
for index, value in enumerate(list)
```

`list` is of type *list*, `index` is a variable representing the index of an item, and `value` is a variable representing the corresponding item. Let us look at a concrete example.

````python
a=["orange","apple","banana","grape","apple","tomato"]
b=[[index,value] for index, value in enumerate(a)]
print(b)
````

The 2nd line uses list comprehension notation to define the list `b`, and the `enumerate()` function is written in the parentheses.

* `b` is a list that is created in the form [index, value]. (That is, b is a "list of lists".) 
* Each of the indices and values are extracted from list `a` using the notation below.

```
for index, value in enumerate(a)
```

List `a` has six items from `a[0]="orange"` to `a[5]="tomato"`, and for each of them, a pair of index `x` and value `a[x]` is created.
* As a result, a total of six pairs are generated from list `a`, from [0, "orange"] to [5, "tomato"]. The whole set is list `b`.

Based on this, let us consider how to obtain all the indices whose value is "apple" in list `a`.
To do so, we can create a list with indices that are extracted from `a` when the value is `apple`.
It can be written as follows.

````python
a=["orange","apple","banana","grape","apple","tomato"]
b=[index for index, value in enumerate(a) if value=="apple"]
print(b)
````

If you run the program and look at the result of line 3, you will see that `b` is the list of all indices k such that a[k]="apple" in the list `a` (i.e., 1 and 4).

* The list comprehension notation in the 2nd line tells us that `b` is a list of indices.
* `index, value` are extracted sequentially from list `a` by the `enumerate()` function, and they are passed through the `if` statement written afterwards.

```
if value=="apple"
```

This statement means "if `value` is equal to `apple`".
* In other words, the whole statement has the following meaning.
* For `index, value` extracted from list `a`, the list `b` contains only the indices where `value` is equal to `apple`.

Therefrore that `b` is a list of all indices k such that `a[k]="apple"`.

## 4.4.3 Deleting values from a list

In the Chapter "Programming in Python for beginner (2) Lists and Dictionaries", we explained the operation for deleting specified values from a list. In this section, we will learn about methods to delete all the specified values from a list.
For example, for a list `a` where
```
a=["peach","orange","apple","orange","grape","banana","orange"]
```
we want to delete all occurrences of `"orange"` from the list.
You can use the `remove()` function  to remove the first `a[1]="orange"` found, but if you want to remove all `"orange"`s, you need to do something else.

Let us first look at a simple example.

````python
a=["peach","orange","apple","orange","grape","banana","orange"]
b=[]

for w in a:
  if w!="orange":
    b=b+[w]

print(b)
````

In this example, the items of list `a` that is not `"orange"` are stored in list `b`. The initial state of `b` is an empty list as shown in the 2nd line. In the `for` loop in lines 4 to 6, for each item `w` that belongs to `a`, $\quad$ `w` is added to `b` only when `w` is not `"orange"`.

Another way is to use the list comprehension notation, as shown below:

````python
a=["peach","orange","apple","orange","grape","banana","orange"]
a=[w for w in a if w!="orange"]
print(a)
````

The 2nd line is the part where all occurrences of `"orange"` are removed from `a`.
In the list comprehension notation after `for`
```
for w in a if w!="orange"
``` 
means "each item `w` of list `a` whose value is not `"orange"`".
And
```
[w for w in a if w!="orange"].
```
makes a list using `w` as its items.

**Question 2.** Create a program that displays the largest and smallest numbers among five integers that will be input from the keyboard.

````python
number = []
for i in range(0,5):
    number.append(int(input("Enter #" + str(i+1) + ": ")))

max = number[0]
min = number[0]
for j in range(0, len(number)):
    if number[j] > max:
        max = number[j]
    elif number[j] < min:
        min = number[j]

print("Max is " + str(max) + ", min is " + str(min) + ".")
````

or $a_{(n+1)/2}$ (when $n$ is an odd number).
