# **1. Programming in Python for beginners (1) Variables and Assignment**

When executing a program, it is often necessary to temporarily store some information (e.g. numbers, strings, etc.) somewhere as a “memo”. Therefore, we will first learn the concept of “variables” (which will act as memos mentioned beforehand), and how to use them to store information temporarily in Python. # 1.1 Basics of variables and their applications

The program below is a simple program that when run, will require us to input two integer values from the keyboard, and the program will calculate their sum and product, and then displays the results. Using this program as a subject, we will learn how to use variables.

````python
print("input two numbers: ")
a=int(input())
b=int(input())

sum=a+b
prod=a*b

print(a,"+",b,"=",sum)
print(a,"*",b,"=",prod)
````

(1) The 1st line

```
print("input two numbers: ")
```

is an instruction to print or display the message "input two numbers: " on the screen.
As you can see, the function of `print()` is to print the message written in parentheses () on the screen. The message to be printed should be enclosed in double quotation marks “”.


(2) Before explaining the statements in the 2nd and 3rd line, let us first understand briefly what a "variable" is.

**Variables** are like data storage boxes used to temporarily store various information such as numbers, strings, etc. as "memos" during the process of executing a program. The minimum rules for using variables are as follows.

1.   Each variable should be created with a unique name (variable name). Using the example above, two variables with the names a and b are created in the 2nd and 3rd lines respectively.
2.   Each variable can store some values. Here,**「values」are not limited to numerical values, but can be information like strings**. Also, it is also possible to retrieve and use the values stored in a variable.
3.   However, each variable **can only store one piece of information at a time**. When a new value is stored in a variable, the previously stored value is lost.

The operation to store a value in a variable is called "assignment", and the format of the assignment statement is as follows.

> **variable name = data to be assigned**

This will assign the data on the right side to the variable specified by the variable name on the left side.
Please note the following 2 points regarding the assignment statement.

* The assignment statement uses an equal sign. **However, this does not mean that「the left side and the right side are equal」, but rather「the right side is assigned to the left side」**.
* In the assignment statement, the value on the right side is calculated first. Then the value is assigned to the variable on the left side.

For example, consider the following statement. (`a` is an integer)
```
a = a + 1
```

The 1st point to note is that this expression does not mean「`a` and `a+1` are equal」, but 「assign the value of `a+1` to `a`」. The 2nd point to note is that when you execute this statement, it will work as follows.

* Calculate the value `a+1` of the right side using the current value of the variable `a`. (For example, if `a`=10, `a+1`=11 is calculated. )
* Then, the calculation result is assigned to the variable `a` again. (As a result, the value of `a` is changed to 11.)

In other words, `a=a+1` represents the action "increase the value of variable `a` by 1".

(3) Now, we take a look at the 2nd line.

```
a=int(input())
```

This is an assignment statement to store the value `int(input())` into a variable named `a`. Since the variable `a` is the first variable to appear here, it will be created here and the value specified by `int(input())` on the right side will be stored in it.

The value `int(input())` on the right side is not a specificied value, but something that is not clear.
First,

```
input()
```

is a function to receive some input from the keyboard.  (You can leave the parentheses empty.) The program will wait until some value is input from the keyboard, and will hold the value when an input is completed.
However, even if we enter an integer value such as 12, `input()` will treat it as a string "12". To convert the input into an integer value, we use the function `int()`.
Therefore, in the following statement,

```
a=int(input())
```

the right side has the following operation.

* use `input()` to receive some input from the keyboard, then
* use the function `int()` to convert the input into an integer value

Then, the result is stored in the variable `a`.
The 3rd line

```
b=int(input())
```

is the same as above.

(4) The 4th line is an empty line inserted to make the program easier to read. This empty line is not required. Let's look at the following 5th and 6th line.

```
sum = a+b
prod = a*b
```

These 2 statements are also assignment statements used to store values in variables. In the statement of `sum` (5th line), the value of `a+b` on the right side is calculated and the result is assigned to the variable `sum` on the left side.
At this point, the values entered from the keyboard are stored in `a` and `b` respectively, and the sum `a+b` is calculated using these values. For example, if 12 and 7 are input from the keyboard in sequence, `a` will be 12 and `b` will be 7, and 19(=12+7) will be stored in the variable `sum`.


The same applies to the statement with `prod` (6th line). The only difference is that `+` (which represents sum) is changed to `*` (which represents product).


(5) The 8th and 9th line are the `print()` function seen in (1), but with a more complicated format.



In this statement, for example, if `a`=12 and `b`=7, then (since `sum`=12+7=19),

> 12 + 7 = 19

will be displayed.
The `+` and `=` which are enclosed by " " will be displayed as what they are, but `a`, `b`, and `sum` will be displayed with the values assigned to them.
The same applies to the following statement (9th line ).

# 1.3 Handling of strings

In Python, it is relatively easy to work with string data. In this section, we will introduce some simple ways to handle strings.


## 1.3.1 Length of strings

To obtain the length of strings, the function `len()` is used.

````python
word="Krungthepmahanakhon Amonrattanakosin Mahintharayutthaya Mahadilokphop Noppharatratchathaniburirom Udomratchaniwetmahasathan Amonphimanawatansathit Sakkathattiyawitsanukamprasit"
print(len(word))
````

## 1.3.2 Concatenation of strings

As explained above, the "+" symbol is used to join multiple strings together.
In the following example, three string variables `a`, `b`, and `c` are joined together to form the string `a+b+c`.

````python
a="I was "
b="born to "
c="love you"
print(a+b+c)
````

Addition of strings will result in a concatenation, but "multiplication" of strings is also possible.
Let us look at the following example.

````python
print("abc"*3)
print(4*"def")
print(len("ab"*(-1)))
````

As you can see in this example, multiplying a string by an integer `k` will result in a concatenated string of `k` strings.

However, if you specify 0 or a negative number for `k`, you will get an empty string.
(Note the result of the 3rd line. It shows that the length of `"ab "*(-1)` is 0.)

## 1.3.3 Obtaining a portion of a string

The following example shows how to obtain the k-th character of a string.

````python
word="LetItBe"
print(word[0])

print(word[-1])
````

In this example, `word[0]` points to "L", the first letter of `word="LetItBe"`. Note that the **numbering starts from 0**.
Also, if you specify -1 as the number, like the following example,

```
word[-1]
```

it will point to the last character of the string `word`. (In this case, it will point to the last character in "LetItBe" which is "e" .) Similarly, `word[-2]` can be used to point to the second-to-last character of a string.


````python
word="Bohemian Rhapsody"
print(word[1:5])
print(word[1:-1])

print(word[5:])
print(word[:5])
````

The expression `word[1:5]` used here is called **slice notation**. The notation `[1:5]` after the variable `word` specifies the range from `word[1]` to just before `word[5]`, that is, from `word[1]` to `word[4]`.
(Again, the first character is `word[0]`. `word[1]` is the second character from the beginning.)

In this case, `word="Bohemian Rhapsody"`, so if `word[1:5]` is specified, "ohem" will be extracted.

```
word[:5]
```

If the starting number is omitted, as in the case above, it is interpreted as `word[0:5]`.

```
word[5:]
```

If the end number is omitted, as in the case above, it means "from `word[5]` to the end of the variable `word`".

There are many expressions in slice notation, which are very useful, but you don't need to memorize all of them for now.

## 1.3.4 String replacement

(1) For example, say you want to replace "Michael" with "John" in the sentence "This is Michael". You can write the following code, but it will not work.

````python
word="This is Michael"
word[-7:]="John"
print(word)
````

In the 2nd line, `word[-7:]` is a slice representing the characters starting from the 7th character counting from the end of the string until the end of the string, specifically `"Michael"`. The intention is to replace it with `"John"` by an assignment statement, but this does not work as intended and results in an error. This is because `word[-7:]` is a string `"Michael"`, not a variable, and the operation of assigning `"John"` to it is not valid.

As a solution, we can do the following as an example.

````python
word="This is Michael"
word=word[:8]+"John"
print(word)
````

In this example, `word[:8]` becomes a slice that represents `"This is "`, which is the range from `word[0]` to `word[7]`, cutting off `"Michael"` at the end. By appending `"John"` to it, it will appear as if Michael is replaced by John. However, the disadvantage of this method is that you have to count the number of characters, in this case we have to slice until the 8th character from the front.

(2) In some cases, you may want to replace all the same characters in a string with other characters.
In the following example, the statement `word.replace("X", "a")` in the 2nd line means, replace all "X" in the variable named word with "a". Then, the result is assigned to the variable `word` again. Now, all the "X" in the variable `word` are replaced by "a".

````python
word="TXke it EXsy"
word=word.replace("X","a")
print(word)
````

In the above example, the "X" and "a" can be made up of a string of two or more characters.
In the following example, by replacing "X" with "that" in 2 places in the 2nd line, and replacing "YZ" with "is" in the 3rd line, the variable `word` will change as follows.


> "XYZX" → "thatYZthat" → "that is that"

Applying this to the example in (1) as shown below,

> word=word.replace("Michael","John")

 `"Michael"` will be replaced with `"John"` .



````python
word="XYZX"
word=word.replace("X","that")
word=word.replace("YZ"," is ")
print(word)
````

One application of the function `word.replace` is to perform the operation of deleting specific characters or strings in a string.
In the following example, by replacing all "e"s with "" (an empty string), all "e"s are deleted from the sentence.

````python
word="Here, there and everywhere"
word=word.replace("e","")
print(word)
````

(3) In some cases, you may want to insert another character or string into a specific place of a string.
This is not an easy task, but requires a little ingenuity. In the following example, we want to insert the relative pronoun "which" in between "book" and "I".

````python
word="This is the book I bought at Junkudo yesterday."
pos=word.find("I")
print(pos)
word=word[:pos]+"which "+word[pos:]
print(word)
````

First, the right side of the 2nd line

```
word.find("I")
```
represents the number of characters in the string `word` where the letter "I" is first found. (The first character is counted as 0. That is, it returns the smallest number k such that `word[k]="I"`. In this case, 17.)
Then, the right side of the 3rd line is constructed as follows.

* `word[:pos]` is a slice notation that represents a string consisting of `word[0]` up to just before `word[:pos]`, specifically "This is a book ".
* Next, by adding `"which "`, the string "which " will be inserted after "This is a book ".
* Finally, `word[pos:]` is a slice that refers to the part "I bought at Junkudo yesterday". By adding this to "which ", the result will show that the string "which" is inserted between "book" and "I".

In conclusion, the program will insert strings using the following procedure.

1. From the 2nd line, get the*「inserting point」*.
2. From the 3rd line, use slice method to split the variable word into 2 parts (*before* inserting point and *after* inserting point), then insert the string you want between these 2 parts.


# 1.4 Additional details regarding data types.

## 1.4.1 Data types in Python

Programs can handle various types of data, and the type of data is called a **data type**. In Python, for example, there are data types for numerical data, which are the「integer type」and the「floating point type」that can handle numbers after the decimal point, and the「complex number type」that can handle complex numbers.
Other than numbers, there are various other data types such as「string type」,「boolean type」that represents logical values of "true/false",「list type」and「dictionary type」which we will learn later.

If you want to know the type of data, use the `type()` function, as shown in the following example.

````python
x=3
y=4-1j
print(type(x))
print(type(y))
print(type(x*y))
````

In this example, `x` and `y` are defined as variables of type integer and type complex respectively. (j is an imaginary unit. However, even in the case of y=4-j, we need to write `y=4-1j` without omitting the factor 1 of j as shown in the 2nd line)

In the 3rd and 4th line, the data types of `x` and `y` are displayed respectively. Make sure that the data types are displayed as `"class int"` (type integer) and `"class complex"` (type complex number).
In the 5th line, the data type of `x*y` (=12-3j) is displayed, which is also "class complex".

````python
a="string"
print(type(a))
a=1
print(type(a))
a=2.0
print(type(a))
````

In this example, the variable `a` is first created as a string variable in the 1st line, and is printed as `"class str"` (type string) in the 2nd line.

Then, in the 3rd line, the assignment `a=1` is made, and the data type of `a` is changed to type integer. Please make sure that the 4th line displays `"class int"` (integer type).

Then, in the 5th line, the assignment `a=2.0` is made. Mathematically, 2.0 is the same as the integer 2, but by adding ".0", Python assumes the value is the floating point number 2.0, not the integer 2. Make sure that the 6th line displays `"class float"` (floating point).

## 1.4.2 Computing data with different data types

For example, try to run the following program.

````python
a=2.0
b=3
print("type of a =",type(a))
print("type of b =",type(b))
print(a+b,type(a+b))
````

In this example, the variables `a` and `b` are both mathematically integers, but as data types in Python, `a` is of type floating point and `b` is of type integer, so they have different data types. (Check the results displayed in the 3rd and 4th line.)

Since they are both numerical data, the operation `a+b` itself is possible, but the question is what happens to the data type of the result ?

In the 5th line, the value of `a+b` and its data type are printed out, but the data type is displayed as `"class float"` (type floating point). In other words, the result of arithmetic between type integer and type floating point is treated as type floating point, which has a higher precision.

What happens if `a` is not treated as type floating point, but as type integer? Let us look at the following example.

````python
a=int(2.71)
b=3
print("type of a =",type(a))
print("type of b =",type(b))
print(a+b,type(a+b))
````

In this case, in the first line the function `int()` is used to define `a`, and this makes it clear that the value on the right side is an integer. (In this case, the ".71" after the decimal point is ignored, and only the integer part 2 remains, which is assigned to `a`. Thus, the value of `a` is the integer type 2.)

Then, in the 2nd line, the variable `b` is also created as a variable with type integer.
In other words, both `a` and `b` are of type integer. (Check the results displayed in the 3rd and 4th line.)

Therefore, the data type of `a+b` is also naturally integer. Check the result of the 5th line.

````python
a=3
b=2
print("type of a =",type(a))
print("type of b =",type(b))
print(a/b,type(a/b))
````

In the example above, both `a` and `b` are defined as type integer variables, but the result of the division `a/b` is mathematically 1.5, which is not an integer.
In such a case, Python automatically calculates `a/b` as a floating-point number. Check the result displayed in the 5th line.

The examples shown above are for numerical operations, but the next example is a mixed operation dealing with numerical data type and string data type. The following example will result in an error.

````python
name="Donald"
age=74
message=name+" is "+age+" years old."
print(message)
````

The error occurs in the 3rd line where the variable `message` is defined.
The variable `name`, `"is"`, and `"years old"` are all strings, but the variable `age` is a number. Since the types of the data to be connected by the `+` sign do not match, the meaning of the `+` sign is not determined, and an error occurs.


````python
name="Donald"
age=74
message=name+" is "+str(age)+" years old."
print(message)
print(name, age)
````

On the other hand, in this program, the variable `age` is enclosed by the **`str()` function** in the 3rd line.
`str()` is a function that converts a given value into data type with type string.
In this way, `age` is also recognized as a string "74", and the symbol `+` is treated as a concatenation of strings.

Similarly, the following statement will also fail due to the mismatched data types.

```

```

Here, `bool()` is used to determine whether the condition in parentheses is "True" or "False".
The inequality `"a">3` compares the letter `a` with the number 3, but an error occurs because "letters and numbers cannot be compared".

However, the following statement does not cause an error.
(The output is True.) 

```

```

Indeed strings can be compared with each other.

## 1.4.3 Floating point error in Python

Try to run the following program.

````python
a=1
b=3.14
c=a+b
print(len(str(c)))
````

In this example, `a` is of type integer, `b` is of type floating point, and `c=a+b` is automatically computed as a floating-point type. Mathematically, a+b=4.14, but if you look closely at the result, you will see a 1 at the end of 4.1400000 for some reason.
This is an error caused by a small difference between the representation of decimals using binary representation and decimal representation in the computer. Although it is a very small error, it is certain that an error is occurring.

Errors are inevitable in numerical computation.
In Python (and other programming languages), when working with floating point numbers on a computer, it is important to take into account **the number of significant digits, that is, the number of digits after the decimal point that are meaningful and reliable**.
