## Question 4
Write a <u>recursive</u> program that implements the <i>expansion along a line formula</i> to compute the determinant of a matrix. Remember that the formula along the line $i$ is $$\det A=\left|\begin{smallmatrix} a_{11} & \cdots & \cdot & \cdots & a_{1n} \\
 \vdots & \ddots & \vdots & \ddots&  \vdots \\ 
a_{i1} & \cdots & a_{ij} & \cdots & a_{in}\\
\vdots & \ddots & \vdots & \ddots & \vdots \\ 
a_{n1} & \cdots & \cdot& \cdots & a_{nn} \end{smallmatrix}\right|
=\sum_{j=1}^n (-1)^{i+j} \, a_{ij}\, \det(\widehat{A}^{ij}),$$ where $\widehat{A}^{ij}$ is the $(ij)$-minor matrix of $A$ (that is, line $i$ and column $j$ are deleted)

```python
A=[[1]]
B=[[1,1],[1,2]]
C=[[1,2,3],[4,5,6],[-7,8,-9]]  # matrix test
customTestCase = [[2,3,2,1],[1,4,3,2],[3,4,1,2],[2,3,1,4]]

def det(matrix, i): # Wise to use i=0 in the recursive calls (not at the main call)
    size = len(matrix) # since determinants only exist for nxn matrices, we can generalize the size of rows and columns.
    # the determinant of a 1x1 matrix is equal to the only entry of the matrix.
    if size == 1:
        return matrix[i][0]
    else:
        # now to handle all other cases.
        determinant = 0
        for j in range(len(matrix[i])):
            # for each a(ij), we want to construct its minor, and recursively look for its determinant.
            minor = [x[:] for x in matrix]
            del minor[i]
            for k in range(len(minor)):
                del minor[k][j]
            # and now for the magic formula: recursively call the determinant of the minor, and sum the products
            # of the minor and the element with the negative/positive sign according to ij
            # to get the final determinant.
            determinant = determinant + ((-1)**(i+j))*matrix[i][j]*det(minor, 0) # using i = 0 for recursive calls
        return determinant

print("The determinant is " + str(det(customTestCase, 0)))
print("The determinant is " + str(det(customTestCase, 1)))
print("The determinant is " + str(det(customTestCase, 2)))
print("The determinant is " + str(det(customTestCase, 3)))
# if the function is built correctly, regardless of which value we pick for i, the determinant will not change.
```
