## Question 6 (MergeSort, requires Question 5 "Merge")
To sort a list L, design a <u>recursive function</u> that cuts the list in the middle, calls recursively the algorithm on the two sublists to sort them, and finally use "Merge" of Question 9  to merge the two sorted lists.
<u>Imposed form</u>
* The function takes as input the list to sort.
* The function returns the list sorted increasingly.


```python
def mergeSort(L):
    if len(L) > 1:

        # Create sub_array2 ← A[start..mid] and sub_array2 ← A[mid+1..end]
        mid = len(L)//2
        sub_array1 = L[:mid]
        sub_array2 = L[mid:]

        # Sort the two halves
        mergeSort(sub_array1)
        mergeSort(sub_array2)
        
        # Initial values for pointers that we use to keep track of where we are in each array
        i = j = k = 0

    # Until we reach the end of either start or end, pick larger among
    # elements start and end and place them in the correct position in the sorted array
        while i < len(sub_array1) and j < len(sub_array2):
            if sub_array1[i] < sub_array2[j]:
                L[k] = sub_array1[i]
                i += 1
            else:
                L[k] = sub_array2[j]
                j += 1
            k += 1

    # When all elements are traversed in either arr1 or arr2,
    # pick up the remaining elements and put in sorted array
        while i < len(sub_array1):
            L[k] = sub_array1[i]
            i += 1
            k += 1

        while j < len(sub_array2):
            L[k] = sub_array2[j]
            j += 1
            k += 1

L = [10, 9, 2, 4, 6, 13]
mergeSort(L)
print(L)
```

## Question 8
Consider `n` points in the plane, represented by a list of their $x$ and $y$-coordinates.

Write a program that find the maximal number of points on a same line (arbitrary line, including vertical).

Example:

* input: `[[1,1],[2,2],[3,3]]`, output: `3`



* input:`[[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]]`, output: `4`



```python
# My approach is to iterate on combinations of 2 points
# in the list, finding the equation (y = mx + c) and then
# checking if the other points are on the line defined by that
# equation.

ex1 = [[1,1],[2,2],[3,3]]
ex2 = [[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]]

L = ex2

maxCohabitors = 0
for i in L:
    for j in L:
        if j != i:
            deltaY = i[1] - j[1]
            deltaX = i[0] - j[0]

            if deltaX != 0 and deltaY != 0: # in the case the line is not horizontal or vertical
                m = deltaY/deltaX

                # now we test the gradient on one of the points
                # to find c

                # y = mx + c, so c = y - mx
                c = i[1]-m*i[0]

                cohabitors = 2 # start at two because we're ssearching for points on the lind OTHER THAN the first two points
                for k in L:
                    if k != i and k != j:
                        if k[1] == m*k[0] + c: # if point k is on the line occupied by both i and j
                            cohabitors += 1
            
            else:
                cohabitors = 2
                # for horizontal lines, we check if points are on the same y coordinate.
                # for vertical lines, we check if points are on the same x coordinate.
                if deltaY == 0:
                    for k in L:
                        if k != i and k != j:
                            if k[1] == i[1]:
                                cohabitors += 1
                
                elif deltaX == 0:
                    for k in L:
                        if k != i and k != j:
                            if k[0] == i[0]:
                                cohabitors += 1
            if cohabitors > maxCohabitors:
                maxCohabitors = cohabitors

print("the maximum number of cohabiting points on an arbitrary line is " + str(maxCohabitors) + ".")
```
