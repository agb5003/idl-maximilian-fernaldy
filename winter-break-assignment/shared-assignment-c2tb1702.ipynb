## Question 1 
Let A be a matrix (seen as a list of lists=list of columns).

**(1.1)**  Write a program that finds the max of the absolute value of its entries.

**(1.2)** And one that computes the list of averages over the entries of each row.

<i> You can test with the matrices A, B and C given in preamble.</i>

```python
A=[[1]]
B=[[1,1],[1,2]]
C=[[1,2,3],[4,5,6],[-7,8,-9]]
matrices = [A, B, C]

test = matrices[ord(input("Which matrix: (use capital letters)")) - 65]

maxval = 0
for i in range(len(test)):
    for j in range(len(test[i])):
        if test[i][j] < 0:
            absval = -test[i][j]
        else:
            absval = test[i][j]
        if absval > maxval or absval > maxval:
            maxval = absval

print(maxval)
```

## Question 5 (Merge)
Given two lists of integers sorted *increasingly*  L and R, write a program that puts all elements of R in L (merge R into L) by maintaining the sequence L sorted. (The list R has less and less elements  and is empty at the end, while L receives the elements of R hence has more and more elements).

<u>Imposed form</u>
* Use the `pop()` command to take the largest (=last) element of R (to be put it in L).
* Use a while loop, taking the condition "R is not empty".
* Needless to say, do not use any kind of sorting functions already existing in Python.

<u> *Hint*:</u> You can start by inserting one element to L and generalize.

```python
L=[-3, -1, 4, 7, 11, 19, 30, 100] # sorted increasingly
R=[10, 20] # sorted increasingly
## After the program we should have:
# L=[-3, -1, 4, 7, 10, 11, 19, 20, 30, 100]
# R=[]

while R!=[]:
    popped=R.pop(-1)
    if popped>L[-1]:
        L = L+[popped]
    else:
        i=0
        while popped>L[i]:
            i+=1
        L[i:i]=[popped]
        
print("L="+str(L))
print("R="+str(R))
```

## Question 7

For a list of non-negative integers, find the largest integer that is made from the non-negative integers. Examples:

* input: `[20,5]`,  output should be `520`

* input: `[6,90,3]`, output should be `9063`
* input `[17,83,7]`, output should be `83717`

Remark: it is fine to manipulate strings instead of integers.

```python
L = [17, 83, 7]
R = []
k = 0

for i in range(len(L)):
    L[i]=str(L[i])
while L!=[]:
    maximum=L[0]
    count=0
    for j in range(len(L)):
        if int(L[j][0])>int(maximum[0]):
            maximum=L[j]
            count=j
        elif int(L[j][0])==int(maximum[0]):
            if int(L[j])>int(maximum):
                maximum=L[j]
                count=j
    R[k:k]=[L.pop(count)]
    k=k+1
output=""
for z in R:
  output=output+z
print(str(output))
```

## Question 8
Consider `n` points in the plane, represented by a list of their $x$ and $y$-coordinates.

Write a program that find the maximal number of points on a same line (arbitrary line, including vertical).

Example:

* input: `[[1,1],[2,2],[3,3]]`, output: `3`



* input:`[[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]]`, output: `4`



```python
eg1 = [[1,1],[2,2],[3,3]]
eg2 = [[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]]

L = eg2

maximum = 0
for p1 in L:
    for p2 in L:
        if p2!=p1:
            dY=p1[1]-p2[1]
            dX=p1[0]-p2[0]
            sameline=2
            for p3 in L:
                if p3!=p1 and p3!=p2:
                    if dX!=0 and dY!=0:
                        m=dY/dX
                        c=p1[1]-m*p1[0]
                        if p3[1]==m*p3[0]+c:
                            sameline+=1
                    else:
                        if dX==0:
                            if p3[0]==p1[0]:
                                sameline+=1
                        elif dY==0:
                            if p3[1]==p1[1]:
                                sameline+=1
            if sameline>maximum:
                maximum=sameline
print(maximum)
```
